```python
titanic_train = pd.read_csv("G:/LearningData/titanic_train.csv")
print(titanic_train.shape)
titanic_train.head()
```

```python
age = titanic_train["Age"]
# print(age.loc[0:10])
age_is_null = pd.isnull(age)
# print(age_is_null)
age_null_true = age[age_is_null]
print(len(age_null_true))
print(age_null_true)
```

```python
mean_age = sum(titanic_train["Age"])/len(titanic_train["Age"])
print(mean_age)
```

```python
good_ages = titanic_train["Age"][age_is_null == False]
correct_mean_age = sum(good_ages)/len(good_ages)
print(correct_mean_age)
```

```python
correct_mean_age = titanic_train["Age"].mean()
print(correct_mean_age)
```

```python
#mean fare for each class
passenger_classes = [1, 2, 3]
fares_by_class = {}
for this_class in passenger_classes:
    pclass_rows = titanic_train[titanic_train["Pclass"] == this_class]
    pclass_fares = pclass_rows["Fare"]
    fare_for_class = pclass_fares.mean()
    fares_by_class[this_class] = fare_for_class
print(fares_by_class)
```

```python
def hundredth_row(column):
    hundredth_item = column.loc[99]
    return hundredth_item

hund_row = titanic_train.apply(hundredth_row)
print(hund_row)
```

```python
def not_null_count(column):
    column_null = pd.isnull(column)
    null = column[column_null]
    return len(null)
```

```python
column_null_count = titanic_train.apply(not_null_count)
print(column_null_count)
```

```python
def which_class(row):
    pclass = row['Pclass']
    if pd.isnull(pclass):
        return 'Unknow'
    elif pclass == 1:
        return 'First Class'
    elif pclass == 2:
        return 'Second Class'
    elif pclass == 3:
        return 'Third Class'
    
classes = titanic_train.apply(which_class, axis=1)
print(classes)
```

```python
def generate_age_label(column):
    if pd.isnull(column['Age']):
        return 'Unknow'
    elif column['Age'] < 18:
        return 'minor'
    elif column['Age'] >= 18:
        return 'adult'
    
age_label = titanic_train.apply(generate_age_label, axis=1)
print(age_label)
```

```python
titanic_train['age_label'] = age_label
age_label_survival_rate = titanic_train.pivot_table(index='age_label', values='Survived', aggfunc=np.mean)
print(age_label_survival_rate)
```
